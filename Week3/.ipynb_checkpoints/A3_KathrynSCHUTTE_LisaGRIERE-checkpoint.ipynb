## Problem statement

You are working as a Data Scientist at a Finance company which proposes home loans. When a customer applies for a home loan, the company studies his demand to decide whether he/she is eligible or not.

Your boss asks you to develop a decision-aid tool to automate the loan eligibility process. These information include customer's personal citeria such as Gender, Marital Status, Education, Number of Dependents, Income, Loan Amount, Credit History, etc... To develop the algorithm/model to be applied in the tool, you are given a data set containing historical information about applicants (**features**) as well as their loan status (**output** given as binary variable indicating whether or not a the loan was approved).
Here below the description of the dataset:

1- Load data

```python
# imports
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# show plots in the notebook
%matplotlib inline
```

```python
# load dataset
df = pd.read_csv('loan_prediction.csv')
# shape of df
print(df.shape)
# head of df
df.head()
```

2- Remove the "Loan_ID" column, and then transform the "Loan_Status" feature from categorical into numerical values (Y ==> 1, N ==> 0)

```python
# your code here
#remove column
df.drop('Loan_ID', axis=1, inplace=True)
#change loan status to numerical values
df['Loan_Status'] = pd.Categorical(df['Loan_Status']).codes
df.head()
```

## Data Exploration

3- For each column, print the number of null values and the number of unique values. If the number of unique values of a given column is less than 10, then print these values.

```python
for col in df.columns:
    print('Column: '+col)
    nb_null = df[col].isnull().sum()
    print('\t Nb of null values = '+ str(nb_null))
    nb_unique = df[col].nunique()
    unique_vals = df[col].unique()
    if nb_unique < 10:
        print('\t Unique values: ' +str(unique_vals))
    else:
        print('\t Nb of unique val: '+str(nb_unique))
        
```

4- Fill the null values in the column **[Married]** by the most frequent one (which is "Yes"). Then, remove all the rows containing at least one NullValue.

```python
# your code here
df['Married'].fillna('Yes', inplace=True)
#drop rows that have any null value
df=df.dropna()
print(df.shape)
```

# Model Training & Evaluation

6- Transform all the text columns of df into numerical ones, and print the head of df

```python
# your code here
for col in df.columns: # Loop through all columns in predictors
    if df[col].dtype == 'object':  # check if column's type is object (text)
        df[col] = pd.Categorical(df[col]).codes  # convert text to numerical
```

```python
# head of df
# your code here
df.head()
```

7-1 Load the target variable "Loan_Status" into a dataframe and call it y. Then, load the features into a second dataframe and call it X. 

```python
y = df['Loan_Status']
X = df.drop('Loan_Status', axis=1, inplace =False)
print (X.columns)
```

7-2 Split the data set into a training set and a test set. Set: test_size=0.3, random_state=123

```python
# your code here
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
print ("train shape", X_train.shape, y_train.shape)
print ("test shape", X_test.shape, y_test.shape)
```

8- Fit a logistic model to the training set, and then report the training and testing accuracies

```python
# your code here
from sklearn.linear_model import LogisticRegression
lrf = LogisticRegression(random_state = 123)
lrf.fit(X_train, y_train)
train_score = lrf.score(X_train, y_train)
test_score = lrf.score(X_test, y_test)
print ('train accuracy =', train_score)
print ('test accuracy =', test_score)
```

9- Print the confusion matrix and compute the **precision** for class **0**, **recall** for class **1** using the formulas learned in class (you should not use metrics.classification_report)

```python
# your code here
from sklearn import metrics
y_pred = lrf.predict(X_test)
C= metrics.confusion_matrix(y_test, y_pred)
print(C)
```

```python
precision_0 =C[0,0]/(C[0,0]+C[1,0])
recall_1 = C[1,1]/(C[1,1]+C[1,0])
print ('precision_0 =', precision_0)
print ('recall_1 =',recall_1)
```

10- Use 5-fold cross-validation and print the mean and standard deviation of scores, how is the mean accuracy compared to the first one with train_test_split

```python
# evaluate the model using 5-fold cross-validation
# your code here
from sklearn import cross_validation
clf = LogisticRegression(random_state=123, fit_intercept = False)
scores = cross_validation.cross_val_score(clf, X, y, scoring = 'accuracy', cv=5)
print (scores)
print ('scores mean =', scores.mean(), 'scores std =', scores.std())
```

Looks good. It's still performing at the same range of accuracy.

The mean accuracy (0.81) is a little lower than the one with train_test_split (0.86)
